## Partielle Ableitung (partial derivative)

Partielle Ableitungen werden verwendet, um Funktionen mit mehreren Variablen abzuleiten. Dabei werden, wenn es sich nur um eine Funktion handelt, Vektoren verwendet. (Hinweis: wenn es sich um eine partielle Ableitung handelt, wird statt diesem $d$ dieses stylisierte $\partial$ verwendet). Für die Funktion $f(x,y) = 3x^2y$ ergibt sich folgende Vollständige Ableitung:
$$\nabla{}f(x,y) = [\frac{\partial{}f(x,y)}{\partial{}x}, \frac{\partial{}f(x,y)}{\partial{}y}] = [6yx, 3x^2]$$
Der daraus entstehende Vektor wird *Gradient* genannt. Der Gradient einer Funktion ist ein Vektor seiner partiellen Ableitungen.

## Matrix Berechnung

Wenn wir statt nur einer Funktion mehere Funktionen ableiten bewegen wir uns von der Vektorrechnung zur Matrizen-Rechnung über. Zur oben bereits abgeleiteten Funktion werden wir nun noch folgende hinzufügen: $g(x,y) = 2x + y^8$. Der Gradient der Funktion ist folgender: $\nabla{}g(x, y) = [2, 8y^7]$.
Wenn man 2 Funktionen hat kann man ihre gradienten in einer matrix stacken. Die Matrix wird dann *Jacobian matrix* oder nur *Jacobian* genannt. Für die Funktion $f$ und $g$ sieht sie folgendermaßen aus:
$$J = \begin{bmatrix}\nabla{}f(x, y)\\ \nabla{}g(x, y)\end{bmatrix} = \begin{bmatrix}\frac{\partial{}f(x,y)}{\partial{}x} & \frac{\partial{}f(x,y)}{\partial{}y}\\ \frac{\partial{}g(x,y)}{\partial{}x} & \frac{\partial{}g(x,y)}{\partial{}y} \end{bmatrix} = \begin{bmatrix}6yx & 3x^2\\ 2 & 8y^7\end{bmatrix}$$
